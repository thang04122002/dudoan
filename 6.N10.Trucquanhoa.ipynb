```python
#loại ouline
import numpy as np
import pandas as pd

df = pd.read_csv('Invistico_Airline.csv')
df.head().T
```

```python
#thong tin kiểu dữ liệu
df.info()
```

```python
#các thống kê cơ bản
df.describe().T
```

```python
#xóa hàng có chưá giá trị missing
df = df.dropna()
df.info()
```

```python
import matplotlib.pyplot as plt
gender_counts = df['Gender'].value_counts()
# Vẽ biểu đồ cột
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color=['blue', 'orange'])
plt.title('Số lượng khách hàng theo giới tính')
plt.xlabel('Giới tính')
plt.ylabel('Số lượng')
plt.xticks(rotation=0)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Gender', hue='satisfaction', palette='pastel')
plt.title('Sự hài lòng theo giới tính của khách hàng')
plt.xlabel('Giới tính')
plt.ylabel('Số lượng')
plt.legend(title='Sự hài lòng')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Customer Type', hue='satisfaction', palette='pastel')
plt.title('Sự hài lòng theo loại khách hàng')
plt.xlabel('Loại khách hàng')
plt.ylabel('Số lượng')
plt.legend(title='Sự hài lòng')
plt.show()
```

```python
# Tổng hợp số liệu hài lòng theo hạng vé
summary = df.groupby(['Class', 'satisfaction']).size().reset_index(name='Count')

# Tạo biểu đồ cột nhóm
plt.figure(figsize=(10, 6))
sns.barplot(data=summary, x='Class', y='Count', hue='satisfaction', palette='muted')

# Thêm tiêu đề và nhãn
plt.title('Sự hài lòng theo hạng vé')
plt.xlabel('Hạng vé')
plt.ylabel('Số lượng khách hàng')
plt.legend(title='Mức độ hài lòng')

# Hiển thị biểu đồ
plt.show()
```

```python

# Chuyển đổi biến mục tiêu 'satisfaction' thành giá trị số
df['satisfaction'] = df['satisfaction'].map({'satisfied': 1, 'dissatisfied': 0})

# Chuyển đổi các biến phân loại thành giá trị số
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})
df['Customer Type'] = df['Customer Type'].map({'Loyal Customer': 1, 'Disloyal Customer': 0})
df['Type of Travel'] = df['Type of Travel'].map({'Business Travel': 1, 'Personal Travel': 0})
df['Class'] = df['Class'].map({'Business': 2, 'Eco Plus': 1, 'Eco': 0})

# Tính toán ma trận tương quan
correlation_matrix = df.corr()

# Hiển thị mối tương quan giữa các biến đầu vào và biến mục tiêu 'satisfaction'
correlation_with_target = correlation_matrix['satisfaction'].sort_values(ascending=False)
print(correlation_with_target)
```

```python
import matplotlib.pyplot as plt

correlation_with_target.drop('satisfaction').plot(kind='bar', figsize=(12, 8))
plt.title('Correlation with Customer Satisfaction')
plt.xlabel('Features')
plt.ylabel('Correlation')
plt.show()
```
