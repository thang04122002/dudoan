# Loại bỏ outline 

```python
numeric_columns = df.select_dtypes(include=['number']).columns
```

```python
### 1. Kiểm tra Outliers bằng IQR
# Tính Q1 (25th percentile) và Q3 (75th percentile)
Q1 = df[numeric_columns].quantile(0.25)
Q3 = df[numeric_columns].quantile(0.75)
IQR = Q3 - Q1

# Xác định các dòng có outliers dựa trên IQR
outliers_iqr_condition = (df[numeric_columns] < (Q1 - 1.5 * IQR)) | (df[numeric_columns] > (Q3 + 1.5 * IQR))
outliers_iqr_count = outliers_iqr_condition.any(axis=1).sum()

### 2. Kiểm tra Outliers bằng Z-score
from scipy import stats
# Tính Z-score
z_scores = np.abs(stats.zscore(df[numeric_columns]))

# Xác định các dòng có outliers dựa trên Z-score (Z-score > 3)
outliers_z_condition = (z_scores > 3)
outliers_z_count = outliers_z_condition.any(axis=1).sum()

# Hiển thị kết quả
print(f'Số lượng dòng có outliers (theo IQR): {outliers_iqr_count}')
print(f'Số lượng dòng có outliers (theo Z-score): {outliers_z_count}')
```

```python
df_cleaned = df[(z_scores <= 3).all(axis=1)]
print('Dữ liệu ban đầu: ',df.shape)
print('Dữ liệu bị loại bỏ: ',df.shape[0] - df_cleaned.shape[0])
print('Dữ liệu còn lại: ',df_cleaned.shape)
```

# Giảm chiều dữ liệu

```python
df_cleaned['Target'] = df_cleaned['satisfaction'].apply(lambda x: 1 if x=='satisfied' else 0)
```

```python
# num_features chứa các features có kiểu dữ liệu là mumber
num_features = df_cleaned.select_dtypes('number').columns

# Lưu các cột có kiểu dữ liệu là object
ob_features = df_cleaned.select_dtypes(include = 'object').columns
```

```python
# Xem biểu đồ tương quan giữa các features numner
import matplotlib.pyplot as plt
import seaborn as sns 

plt.figure(figsize = (18, 11))

sns.heatmap(df_cleaned[num_features].corr(),annot=True, cmap='coolwarm', linewidths=0.5)

plt.title('Correlation Heatmap')
```

```python
# # Xem biểu đồ tương quan giữa các features object
from dython.nominal import associations

ob_features = df_cleaned.select_dtypes(include = 'object').columns

# Tính toán ma trận tương quan cho các biến phân loại
assoc = associations(df_cleaned[ob_features], nominal_columns='all', plot=False)

# Lấy ma trận tương quan từ kết quả
corr_matrix = assoc['corr']

# Hiển thị biểu đồ heatmap với các giá trị hệ số tương quan trong từng ô
plt.figure(figsize=(10, 7))
sns.heatmap(corr_matrix, annot=True, linewidths=0.5)
```

```python
# Lựa chọn num_features quan trọng
target = 'Target'

num_features = df_cleaned.select_dtypes(include = [np.number]).columns.drop(target)

correlations = df_cleaned[num_features].corrwith(df_cleaned[target]) # tính toán chỉ số tương quan của các features khác so vs target

threshold = 0.1

selected_num_features = correlations[abs(correlations) > threshold].index.tolist() 
```

```python
selected_features = []

selected_features.extend(selected_num_features)

selected_features.extend(ob_features)

print(selected_features)
```

```python
df_cleaned = df_cleaned[selected_features]
```

# Chuyển đổi dữ liệu

```python
from sklearn.preprocessing import LabelEncoder

# Khởi tạo bộ mã hóa nhãn
le = LabelEncoder()

# Áp dụng bộ mã hóa nhãn cho các cột thuộc loại dữ liệu categorical
df_cleaned[ob_features] = df_cleaned[ob_features].apply(lambda col: le.fit_transform(col))
```
